```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
```

```python
Q_a_16_700 = np.array([5.23, 6.00, 6.67, 7.73, 8.24])
Q_a_21_700 = np.array([7.32, 8.29, 9.21, 10.26, 10.53])
Q_a_32_700 = np.array([12.63, 13.99, 15.27, 16.78, 17.86])
Q_a_32_750 = np.array([13.48, 15.27, 16.63, 17.96, 19.23])
Q_a_32_800 = np.array([13.48, 15.27, 16.63, 17.82, 19.23])

Q_w_16_700 = np.array([12.23, 16.65, 18.07, 22.98, 24.44])
Q_w_21_700 = np.array([17.63, 22.69, 25.41, 30.17, 30.60])
Q_w_32_700 = np.array([32.67, 40.18, 52.50, 58.86, 62.66])
Q_w_32_750 = np.array([60.12, 62.85, 71.71, 75.89, 79.50])
Q_w_32_800 = np.array([68.87, 84.78, 92.39, 95.80, 103.64])

R_a_16_700 = np.array([0.43, 0.36, 0.37, 0.34, 0.34])
R_a_21_700 = np.array([0.42, 0.37, 0.36, 0.34, 0.34])
R_a_32_700 = np.array([0.39, 0.35, 0.29, 0.29, 0.29])
R_a_32_750 = np.array([0.22, 0.24, 0.23, 0.24, 0.20])
R_a_32_800 = np.array([0.20, 0.18, 0.18, 0.19, 0.19])
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(Q_a_16_700, Q_w_16_700)
ax.plot(Q_a_21_700, Q_w_21_700)
ax.plot(Q_a_32_700, Q_w_32_700)
ax.grid()
ax.set_ylabel('$Q_ж, 10^5 \cdot м^3/с$')
ax.set_xlabel('$Q_г, 10^5 \cdot м^3/с$')
ax.text(5, 20, r'$d=16 мм$')
ax.text(8, 30, r'$d=21 мм$')
ax.text(13.5, 55, r'$d=32 мм$')
```
